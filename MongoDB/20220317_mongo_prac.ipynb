```python
from pymongo import MongoClient
from modules import mod_sql2
import pandas as pd
```

```python
client = MongoClient(
    host = "localhost",
    port = 27017
)
```

```python
client.list_database_names()        #db리스트 보여줌
```

```python
db = client["ubion"]    #use명령어 역할과 유사
```

```python
db.test.find()          #db.test.find()는 딕셔너리 값 .
list(db.test.find())    #-> 리스트로 변환해서 한번에 보여줌.
```

```python
db.test.find_one()      #제일 위에 하나만 보여줌
```

```python
data = {
    "1" : "1",
    "2" : "2",
    "3" : "3"
}

db.test.insert_one(data)    #"하나의" 딕셔너리 추가
```

```python
list(db.emp.find())
```

```python
list(db.emp.find(
    {
        "ENAME" : "SMITH"
    }                       #where 자리에 ename = smith 이고
                            #col 자리에 아무것도 안씀 -> 모든 열 출력
))
```

```python
list(db.emp.find(
    {},             #where 자리
    {
        "EMPNO" : 1,
        "ENAME" : 1,
        "DEPTNO" : 1,
        "_id" : 0
    }               #col 자리
))
```

```python
list(db.emp.find().sort("ENAME"))   #아무 조건도 안 건 find, + sort함수
```

```python
list(db.emp.find(
        {
            "SAL" : {
                "$gt" : 2450
                }
        }
        
    ))                      #sal이 2450 보다 큰 값 출력
```

```python
list(db.emp.find(
        {
            "SAL" : {
                "$gte" : 2450
                }
        }
        
    ))                      #sal 2450 '이상'
```

```python
list(db.emp.find(
    {
        "$or" : [
            {"ENAME" : "SMITH"},
            {"ENAME" : "ALLEN"}
        ]
    }
))                      #or 조건
```

```python
list(db.emp.find(
    {
        "$and" : [
            {"ENAME" : "SMITH"},
            {"JOB" : "CLERK"}
        ]
    }
))                              #and 조건
```

```python
list(db.emp.find(
    {
            "ENAME" : "SMITH",
            "JOB" : "CLERK"
        
    }
))                      # and는 그냥 조건 두 개 달아도 됨
```

```python
a = db.emp.find(
        {
            "SAL" : {"$gt" : 2000 ,"$lt" : 4000}            
        },
        {
            "_id" : 0
        }
        
    ) 

pd.DataFrame(a)
```

```python
k = db.emp.find(
    {
        "ENAME" : {
            "$regex" : "E"
        }
    },
    {"_id" : 0}
)                               #^(시작), / (끝문자)$, / .(어디든) -> 안써도 됨

pd.DataFrame(k)
```

```python
pd.DataFrame(db.emp.find())
```

```python
q = db.emp.find(
    {
        "SAL" : {"$gt" : 1250 , "$lte" : 3000}, #and조건
        "JOB" : "SALESMAN"
    },
    {
        "_id" : 0,
        "ENAME" : 1,
        "JOB" : 1,
        "SAL" : 1,
        "DEPTNO" : 1
    }
).sort("ENAME")             #sort(기준, 1 or -1/ 오름차순, 내림차순)
pd.DataFrame(q)
```

```python
db.emp.update_one(
    {
        "ENAME" : "SMITH"
    },
    {
        "$set" : {
            "SAL" : 880
        }
    }
)                       # 조건에서 중복되는 데이터가 있어도 하나만 바뀜, one
```

```python
sm = db.emp.find(
    {
        "ENAME" : "SMITH"
    }
)

pd.DataFrame(sm)
```

```python
db.emp.update_one(
    {
        "JOB" : "ANALYST"
    },
    {
        "$set" : {"SAL" : 3300}
    }
)                       # 조건에서 중복되는 데이터가 있어도 하나만 바뀜, one
```

```python
an = db.emp.find(
    {
        "JOB" : "ANALYST"
    }
)

pd.DataFrame(an)        
```

```python
db.emp.update_many(
    {
        "JOB" : "ANALYST"
    },
    {
        "$set" : {"SAL" : 3300}
    }
)                       # 조건에서 겹치는 데이터 모두 변경, many
```

```python
an2 = db.emp.find(
    {
        "JOB" : "ANALYST"
    }
)

pd.DataFrame(an2)
```
