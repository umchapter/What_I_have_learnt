```python
fillna_sql = """
               SELECT
               EMPNO,
               ENAME,
               IFNULL(COMM, 0) AS COMM
               FROM emp
          """                                 #조건문, ifnull 결측값 채우기

result_fillna = db.executeAll(fillna_sql)
pd.DataFrame(result_fillna)
```

```python
isnull_sql = """
               SELECT
               EMPNO,
               ENAME,
               NOT ISNULL(COMM) AS COMM
               FROM emp
             """                                 #조건문, isnull 결측값 판단 null에 1

result_isnull = db.executeAll(isnull_sql)
pd.DataFrame(result_isnull)
```

```python
case_sql = """
            SELECT
            ENAME,
            CASE
                WHEN JOB = "ANALYST" THEN "분석가"
                WHEN JOB = "CLERK" THEN "점원"
                WHEN JOB = "MANAGER" THEN "관리자"
                WHEN JOB = "PRESIDENT" THEN "사장"
                ELSE "판매사원"
            END AS JOB
            FROM emp
            """                         #CASE 함수, case는 end 되어야 한다. when, then 조건, ELSE 가능

result_case = db.executeAll(case_sql)
pd.DataFrame(result_case)
```

```python
join_sql = """
                SELECT 
                emp.ENAME AS "이름",
                CASE
                    WHEN emp.JOB = "ANALYST" THEN "분석가"
                    WHEN emp.JOB = "CLERK" THEN "점원"
                    WHEN emp.JOB = "MANAGER" THEN "관리자"
                    WHEN emp.JOB = "PRESIDENT" THEN "사장"
                    ELSE "판매사원"
                END AS "직급",
                IF(emp.SAL >= 2450, 'HIGH', 'LOW') AS "급여",              
                CONCAT(emp.DEPTNO, '-', 
                                    CASE
                                        WHEN dept.DNAME = "RESEARCH" THEN "연구개발"
                                        WHEN dept.DNAME = "ACCOUNTING" THEN "회계"
                                        WHEN dept.DNAME = "SALES" THEN "영업"
                                        WHEN dept.DNAME = "OPERATIONS" THEN "경영지원"
                                    END
                ) AS "부서번호-부서명",
                dept.LOC AS "위치"
                
                FROM emp 
                
                LEFT JOIN dept 
                
                ON 
                emp.DEPTNO = dept.DEPTNO
            """

result_join = db.executeAll(join_sql)
pd.DataFrame(result_join)
```

```python
#판다스로 구현

select_emp_sql = """
                SELECT * FROM emp
                 """

emp = db.executeAll(select_emp_sql)
dt_emp = pd.DataFrame(emp)      #emp 불러오기

select_dept_sql = """
                    SELECT * FROM dept
                  """

dept = db.executeAll(select_dept_sql)
dt_dept = pd.DataFrame(dept)    #dpt 불러오기

dt_emp2 = dt_emp[["ENAME", "JOB", "SAL", "DEPTNO"]] #필요한 칼럼만 불러오기

dt_join = dt_emp2.join(dt_dept.set_index("DEPTNO"), on="DEPTNO") #join

dt_join["DEPTNO"] = dt_join["DEPTNO"].astype(str) + "-" + dt_join["DNAME"]
                                                #칼럼 병합, astype 주의
dt_join.drop(columns=["DNAME"], inplace=True)   #잉여 칼럼 드랍

dt_join.columns = ["이름", "직급", "급여", "부서번호-부서명", "위치"] #칼럼명 변경

def rename_job(x) :
    if x == "CLERK":
        return "점원"
    elif x == "SALESMAN" :
        return "영업사원"
    elif x == "MANAGER" :
        return "관리직"
    elif x == "ANALYST" :
        return "분석가"
    elif x == "PRESIDENT" :
        return "사장"           # 직급 번역 함수

dt_join["직급"] = dt_join["직급"].apply(rename_job) #apply로 적용

def rename_dname(x) :
    if x[3:] == "RESEARCH" :
        return x.replace("RESEARCH", "연구개발")
    elif x[3:] == "SALES" :
        return x.replace("SALES", "영업")
    elif x[3:] == "ACCOUNTING" :
        return x.replace("ACCOUNTING", "회계")      #부서명 번역 함수
                                                    #indxeing과 replace 응용

dt_join["부서번호-부서명"] = dt_join["부서번호-부서명"].apply(rename_dname) #apply

dt_join


```
