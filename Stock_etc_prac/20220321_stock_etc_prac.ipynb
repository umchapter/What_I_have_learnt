```python
select_sql = """
                SELECT * FROM stock_trimmed
             """

k = db.executeAll(select_sql)
df = pd.DataFrame(k)
df
```

```python
for i in df.columns :
    print(i)
```

```python
df.iloc[:,1:] = df.iloc[:, 1:].apply(lambda x : round(x,3))     #숫자 인덱싱 iloc
df
```

```python
df["Close-TSLA"] - df["Open-TSLA"]
```

```python
col_l = []
def fluc():
    k = []
    
    for i in df.columns :
        for j in df.columns :
            if (i[-4:] == j[-4:]) and (i.startswith("Close") and j.startswith("Open")) :
                k.append(df[i] - df[j])
                col_l.append(f"Fluc-{i[-4:]}")
    return k


fl = fluc()
fl2 = pd.DataFrame(fl).T
fl2.columns = col_l
```

```python
fl2
```

```python
col_l
```

```python
df2 = pd.concat([df, fl2], axis=1)
```

```python
df2
```

```python
col_l2 = []
def fluc_rate():
    k = []
    
    for i in col_l :
        for j in df.columns :
            if (i[-4:] == j[-4:]) and (j.startswith("Open")) :
                k.append((df2[i] / df[j])*100)
                col_l2.append(f"Fluc_Rate-{i[-4:]}")
    return k

fr = fluc_rate()
fr2 = pd.DataFrame(fr).T
fr2.columns = col_l2
fr2
```

```python
df3 = pd.concat([df2, fr2], axis=1)         #concat함수는 [리스트]로 묶어서, axis -> 결합할 축 방향 설정.
df3
```

```python
name_list = ["AAPL", "AMZN", "CPNG", "TSLA"]
plt.figure(figsize=(15,15))             #서브플롯 띄우기 전에 피겨 설정
plt.tight_layout(pad=10, h_pad=10, w_pad=10, rect=(2,2,2,2))    #서브플롯 간 여백 설정
for i in range(1, 17) :    
    if i < 5 :
        plt.subplot(4,4,i)
        plt.title(f"Open-{name_list[i-1]}")
        plt.plot(df3[f"Open-{name_list[i-1]}"], f"C{i-1}")
    elif 5<= i < 9 :
        plt.subplot(4,4,i)
        plt.title(f"Close-{name_list[i-5]}")
        plt.plot(df3[f"Close-{name_list[i-5]}"], f"C{i-5}")
    elif 9<= i < 13 :
        plt.subplot(4,4,i)
        plt.title(f"Fluc-{name_list[i-9]}")
        plt.plot(df3[f"Fluc-{name_list[i-9]}"], f"C{i-9}")
    elif 13<= i < 17 :
        plt.subplot(4,4,i)
        plt.title(f"Fluc_Rate-{name_list[i-13]}")
        plt.plot(df3[f"Fluc_Rate-{name_list[i-13]}"], f"C{i-13}")

plt.show()
```
