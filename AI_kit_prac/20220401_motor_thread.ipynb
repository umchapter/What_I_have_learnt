```python
import RPI.GPIO as GPIO
from time import sleep

# 모터 상태
STOP = 0
GOGO = 1
BACK = 2

# 모터 채널
CH1 = 0
CH2 = 1

# PIN 입출력
OUTPUT = 1
INPUT = 0

# PIN 설정
HIGH = 1
LOW = 0

# PWM PIN
ENA = 26 #37
ENB = 0  #27

#GPIO PIN
IN1 = 19 #37
IN2 = 13 #35
IN3 = 6  #31
IN4 = 5  #29

def setPin(EN, INA, INB) :
    GPIO.setup(EN, GPIO.OUT)
    GPIO.setup(INA, GPIO.OUT)
    GPIO.setup(INB, GPIO.OUT)
    
    # 100 khz PWM을 동작시켜라 / pwm : 아날로그와 유사한 신호
    pwm = GPIO.PWM(EN, 100)
    pwm.start(0)    # 핀설정 끝나고 pwm out
    return pwm

def setMortorControl(pwm, INA, INB, speed, stat) :
    pwm.ChangeDutyCycle(speed)

    if stat == GOGO :
        GPIO.output(INA, HIGH)  # 움직일 쪽을 1로 해서 전압 줌
        GPIO.output(INB, LOW)
    elif stat == BACK :
        GPIO.output(INA, LOW)
        GPIO.output(INB, HIGH)
    elif stat == stop :
        GPIO.output(INA, LOW)
        GPIO.output(INB, LOW)

def setMortor(ch, speed, stat) :
    if ch == CH1 :
        setMortorControl(pwmA, IN1, IN2, speed, stat)
    else :
        setMortorControl(pwmB, IN1, IN2, speed, stat)

GPIO.setmode(GPIO.BCM)      #PIN 설정 끝나면 모드 설정

pwmA = setPin(ENA, IN1, IN2)
pwmB = setPin(ENB, IN3, IN4)

setMortor(CH1, 80, GOGO)
setMortor(CH2, 80, GOGO)
sleep(3)
setMortor(CH1, 40, BACK)
setMortor(CH2, 40, BACK)
sleep(3)
setMortor(CH1, 100, GOGO)
setMortor(CH2, 100, GOGO)
sleep(3)
setMortor(CH1, 0, STOP)
setMortor(CH2, 0, STOP)

GPIO.cleanup()
```
