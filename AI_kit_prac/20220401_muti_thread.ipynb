```python
import time
from threading import Thread

def one() :
    i = 0
    while 1 :
        print(i)
        i += 1
        time.sleep(2)
        if i == 15 :
            break

def two() :
    i = 0
    while 1 :
        i += 1
        print("Yeah")
        time.sleep(3)
        if i == 10 :
            break


th1 = Thread(target= one, args = ())    # 쓰레드1에 함수 one 할당
th2 = Thread(target= two, args = ())    # 쓰레드2에 함수 two 할당

th1.start()     # 병렬 실행
th2.start()
```

```python
# motor_end
from RpiMotorLib import rpi_dc_lib
import RPI.GPIO as GPIO
from threading import Thread
import time

MotorOne = rpi_dc_lib.L298NMDc(19, 13, 26, 50, False, "motor_one")
MotorTwo = rpi_dc_lib.L298NMDc(6, 5, 0, 50, False, "motor_two")

GPIO.setmode(GPIO.BCM)

trig = 17
echo = 27
GPIO.setup(trig, GPIO.OUT)
GPIO.setup(echo, GPIO.IN)

def motor_one_speed(speed) :
    if speed < 0 :
        MotorOne.backward(-speed)
    elif speed > 0 :
        MotorOne.forward(speed)
    else :
        MotorOne.stop(0)

def motor_two_speed(speed) :
    if speed < 0 :
        MotorTwo.backward(-speed)
    elif speed > 0 :
        MotorTwo.forward(speed)
    else :
        MotorTwo.stop(0)

def motor_move() :
    while 1 :
        motor_one_speed(20)
        motor_two_speed(20)
        time.sleep(3)
        motor_one_speed(-20)
        motor_two_speed(-20)
        time.sleep(3)

def ultra() :
    while True :
        GPIO.setup(trig, GPIO.OUT)
        GPIO.setup(echo, GPIO.IN)

        GPIO.output(trig, False)
        time.sleep(0.5) # 0.005 초 대기 / milisec

        GPIO.output(trig, True) # 초음파 쏨
        time.sleep(0.0001)      # 쏘고 아주 잠깐 기다렸다가
        GPIO.output(trig, False)# 초음파 끈다

        while GPIO.input(echo) == 0 :
            pulse_start = time.time()   #신호 없을 때는 현재시간 기록
        while GPIO.input(echo) == 1 :
            pulse_end = time.time()

        pulse_duration = pulse_end - pulse_start
        distance = pulse_duration * 17000
        dstance = round(distance, 2)
        print("Distance is : ", distance, " cm")

th1 = Thread(target= motor_move, args = ())    # 쓰레드1에 함수 one 할당
th2 = Thread(target= ultra, args = ())    # 쓰레드2에 함수 two 할당

th1.start()     # 병렬 실행
th2.start()
```
